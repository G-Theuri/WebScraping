```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv(r"C:\MyProjects\WebScraping\Nuria\children.csv")
data.tail()
```

```python
data = data.drop('currency', axis=1)
data.head()
```

```python
data['book_price'] = data['book_price'].apply(pd.to_numeric, errors='coerce')
data.info()
```

```python
sns.scatterplot(data=data,x='book_price', y='book_category', sizes=(200, 2000), hue='book_category', legend=False, )
plt.title('Price distribution on book category')
plt.show()
```
